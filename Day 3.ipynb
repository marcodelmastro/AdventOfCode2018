```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("./data/input3.txt") as f:
    lines = [ l.rstrip('\n') for l in f]
lines
```

```python
claims = []
for l in lines:
    p = l.split(" ")
    coord = (p[2].replace(":","")).split(",")
    dim = p[3].split("x")
    claims.append((coord,dim))
claims
```

```python
fabric = np.zeros((1000,1000))

for claim in claims:
    coord = claim[0]
    dim = claim[1]
    for x in range(int(coord[0]),int(coord[0])+int(dim[0])):
        for y in range(int(coord[1]),int(coord[1])+int(dim[1])):
            fabric[y][x] += 1
```

```python
fig = plt.figure(figsize=(8,8), dpi=100)
plt.imshow(fabric)
plt.show()
```

```python
fabric[fabric>1] = 2
overlap = int(fabric[fabric==2].sum())//2
print(overlap)
```
